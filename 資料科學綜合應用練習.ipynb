```python
#引入模組，並將其取成 plt,ap,np縮寫
#將csv資料讀入，並使用'big5'解析文字

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams['font.family'] = 'DFKai-SB' #使用中文字體（黑體）, 只適用 Win10

df = pd.read_csv("taipei.csv", encoding = "big5")
```

### 統計分析

```python
#利用'4天候'與'受傷人數來做分析，顯示出表格與長條圖。

print(df.groupby("4天候")['受傷人數'].sum())

#print(pd.value_counts(df['4天候']))
plt.bar(df.groupby("4天候")['受傷人數'].sum().index, df.groupby("4天候")['受傷人數'].sum().values)

plt.show()
```

```python
#利用年齡資料，取出每個年齡有幾起交通事故，年齡從小到大，並將資料與折線圖顯示

print("年齡\t件數")

a = pd.value_counts(df['年齡']).sort_index()

print(a)

plt.plot(a.index, a.values)
```

```python
#利用'區序'與'受傷人數'去做分析，了解每個區的交通事故造成的受傷人數數量，顯現出資料與圓餅圖。

print(df.groupby("區序")['受傷人數'].sum())

b =  df.groupby("區序")['受傷人數'].sum()

plt.figure(figsize = (4, 4))

plt.pie(b.values, labels = b.index)

plt.show()
```

```python
#利用'速限'去做分析，了解在哪個範圍是最常出現交通事故的，並顯示出資料與箱型圖。

print(pd.value_counts(df['7速限']))

print(pd.value_counts(df['7速限']).index)

x = df['7速限'].values

r = plt.boxplot(x, showfliers = True) #show outliers(異常值)

plt.show()
```

```python
#找出每個區的最大受傷人數，並顯示出資料與折線圖。

plt.figure(figsize = (10, 3))

c= df.groupby("區序")

print(c['受傷人數'].max())

plt.plot(c['受傷人數'].max().index, c['受傷人數'].max())
```
